```python
import pandas as pd
```

```python
# Create reference to CSV file
csv_path = "../Resources/purchase_data.csv"

# Import the CSV into a pandas DataFrame
pymoli_df = pd.read_csv(csv_path)
pymoli_df
```

```python
# Count of Total Players 
total_players = len(pymoli_df["SN"].unique())
total_players

player_count = pd.DataFrame({"Total players":total_players},index = [0])
player_count


```

```python
# Finding the # of unique items, average price, # of Purchases, and Total Revenue
unique_items = len(pymoli_df["Item ID"].unique())

average_price = pymoli_df["Price"].mean()

purchases_count = pymoli_df["Purchase ID"].count()

total_revenue = pymoli_df["Price"].sum()
```

```python
# Summarizing the above cell in a DataFrame
purchasing_analysis_df = pd.DataFrame({"Unique Items": unique_items, 
                                       "Average Price": average_price, 
                                       "Number of Purchases": purchases_count, 
                                       "Total Revenue": total_revenue}, index=[0])
purchasing_analysis_df["Average Price"] = purchasing_analysis_df["Average Price"].map("${:.2f}".format)
purchasing_analysis_df["Total Revenue"]= purchasing_analysis_df["Total Revenue"].map("${:.2f}".format)
purchasing_analysis_df
                                     
```

```python
# Group players by gender 
grouped_gender = pymoli_df.groupby("Gender")

unique_genders = grouped_gender["SN"].nunique()

percentage_players = (unique_genders/total_players) * 100
                  
#Summarize the gender information
gender_demographics_df = pd.DataFrame({"Total Count": unique_genders, 
                                       "Percentage of Players": percentage_players})

gender_demographics_df["Percentage of Players"] = gender_demographics_df["Percentage of Players"].map("{:.2f}%".format)
                                     
gender_demographics_df.sort_values(by = ['Percentage of Players'],ascending = False)
```
